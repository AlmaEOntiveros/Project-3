```python
# Import the necessary library
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.widgets import CheckButtons
```

```python
# Load the 2017 data file into a DataFrame
df2017 = pd.read_excel('Resources/Employee Diversity.xlsx', sheet_name='2017')

# Clean data by replacing zeros as NaN
df2017.replace(0, np.nan, inplace=True)

df2017.head(10)
```

```python
# Plotting the bar chart for Female %
plt.figure(figsize=(10, 6))
bars = plt.bar(df2017['Company'], df2017['Female %'], color='lightcoral')
plt.title('2017 Female Employee % by Top 10 Tech Companies')
plt.xlabel('Company')
plt.ylabel('Female %')
plt.xticks(rotation=45)

# Adding percentage values on top of the bars
for bar in bars:
    yval = bar.get_height()
    if not pd.isna(yval):  # Check for NaN values
        plt.text(bar.get_x() + bar.get_width() / 2.0, yval, f'{yval:.0f}%', ha='center', va='bottom')

plt.tight_layout()
plt.savefig("2017 Female % by Company.png")
plt.show()
```

```python
# Plotting the bar chart for Asian %
plt.figure(figsize=(10, 6))
bars = plt.bar(df2017['Company'], df2017['% Asian'], color='lightblue')
plt.title('2017 Asian Employee % by Top 10 Tech Companies')
plt.xlabel('Company')
plt.ylabel('Asian %')
plt.xticks(rotation=45)

# Adding percentage values on top of the bars
for bar in bars:
    yval = bar.get_height()
    if not pd.isna(yval):  # Check for NaN values
        plt.text(bar.get_x() + bar.get_width() / 2.0, yval, f'{yval:.0f}%', ha='center', va='bottom')

plt.tight_layout()
plt.savefig("2017 Asian % by Company.png")
plt.show()
```
